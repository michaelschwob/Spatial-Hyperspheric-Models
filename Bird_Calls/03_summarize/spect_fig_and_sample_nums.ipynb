```python
import os, random 
import librosa 
import matplotlib.pyplot as plt 
import numpy as np 
from glob import glob
from matplotlib import cm 
import pandas as pd
```

### 1. Make sample spectrogram figure.

```python
def get_random_choices(seed, calls):
    
    random.seed(seed)
    random_choices = []
    
    #spects = glob('../01_data/spects_1s/[!noise]*')

    for call in calls:
        call_path = glob(os.path.join('/Users/calzada/birdsongs/wood/01_data', 'spects', call, '*.png'))
        print(call_path)
        random_choices += random.choices(call_path, k = 3)
    
    return random_choices
```

```python
def get_random_choice_wavs(random_choices):
    trimmed_wavs = []
    beg = '/Users/calzada/birdsongs/wood/01_data/clean/clean_'
    end = '.wav'
    
    for file in random_choices:
        call = file.split('/')[7]
        print(call)
        num = file.split('/')[-1].split('_')[-1].split('.')[0]
        trimmed_name = beg + call + '/' + 'trimmed_'+ str(num) + end
        trimmed_wavs.append(trimmed_name)
    
    return trimmed_wavs
```

```python
def create_spect_grid(audio_files, row_labels):

    fig, axs = plt.subplots(3,3, dpi = 600, figsize = (20,10)) #, sharex = True, sharey = True)

    for i, ax in enumerate(axs.flat):
        if i % 3 < 3:        
            if i % 3 == 0:  # Label only the first column of each row
                ax.text(-0.3, 0.5, row_labels[i // 3], fontsize=20, rotation=90, va='center', ha='right', transform=ax.transAxes)
            
            y, sr = librosa.load(audio_files[i], sr = 44100) 
            ms = librosa.feature.melspectrogram(y=y, sr=sr)
            S_db = librosa.power_to_db(ms, ref=np.max)
            img = librosa.display.specshow(S_db, sr=sr, ax = ax, cmap = 'viridis', y_axis = 'mel', x_axis = 'time')
            
    
    fig.subplots_adjust(right=0.85)
    cbar_ax = fig.add_axes([0.88, 0.15, 0.02, 0.7]) # side color bar 
    fig.colorbar(img, cax=cbar_ax, format='%+2.f dB')
    plt.show()
    fig.savefig('spect_samples.jpg')
```

```python
calls = ['laugh', 'drum', 'pik'] # Order of how the calls will be plotted on grid 
rands = get_random_choices(seed = 8888, calls = calls)
print(rands)
random_wavs = get_random_choice_wavs(rands)
```

```python
row_labels = ['Whinny', 'Drum', 'Pik']
```

```python
create_spect_grid(random_wavs, row_labels)
```
